```python
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#Quantidades de linhas e colunas
df.shape
```

```python
#Verificando tipo de dados
df.dtypes
```

```python
#Receber valor da receita total
df["Valor Venda"].sum()
```

```python
#Receber valor da receita total
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])  #Crio a coluna custo
```

```python
df.head()
```

```python
#Qual o custo total
round(df["Custo"].sum(), 2)
```

```python
#Agora que temos receita e custo podemos achar o lucro
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head(1)
```

```python
#Lucro total
round(df["Lucro"].sum(), 2)
```

```python
#Criando coluna com total de dias pra envio do produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head()
```

```python
#Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
df["Tempo_envio"].dtype
```

```python
df.groupby("Marca")["Tempo_envio"].mean()
```

```python
#verificando se tem valores faltando
df.isnull().sum()
```

```python
pd.options.display.float_format = '{:20,.2}'.format
```

```python
#Saber lucro por ano e por marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#Resetando o Index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Grafico de total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Receita")
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro X Ano")
plt.xlabel("Ano")
plt.ylabel("Receita")
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df.head(1)
```

```python
pd.options.display.float_format = '{:20,.2}'.format
```

```python
df.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro X Mês")
plt.xlabel("Marca")
plt.ylabel("Lucro")
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro X Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal')
```

```python
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro X Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal')
```
