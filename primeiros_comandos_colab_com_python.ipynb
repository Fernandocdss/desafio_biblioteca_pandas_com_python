```python
#Criando e manipulando variável do tipo string
frase = "Fernando é o cara"
frase
```

```python
#As strings sao lidas da esquerda pra direita começando pelo índice 0
#Podendo ser acessada da sequinte forma
frase[2]
```

```python
#Podemos tambem acessar tal trecho da frase
frase[0:8]
```

```python
#Coloca os caracteres do indice 0 ate o final da frase
frase[0:]
```

```python
#Coloca os caracteres do inicio da frase até o caractere 8
frase[:8]
```

```python
#Escreve a frase de trás pra frente
frase[::-1]
```

```python
#Conta quantas vezes a variável ou a string aparece na frase
frase.count("a")
```

```python
#Informa a frase no maiúsculo
frase.upper()
```

```python
#Informa a frase no minuscúlo
frase.lower()
```

```python
#Coloca os primeiros caracteres em maiúsculo
frase.title()
```

```python
#Substitui a palavra no primeiro parâmetro pela segunda
frase.replace("cara", "bonitão")
```
